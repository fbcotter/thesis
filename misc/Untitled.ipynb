```python
%matplotlib notebook
import numpy as np
import dtcwt
import matplotlib.pyplot as plt
from pytorch_wavelets import DTCWTForward, DTCWTInverse
import torch
```

```python
xfm = DTCWTForward(J=1, biort='near_sym_a', qshift='qshift_a')
ifm = DTCWTInverse(J=1, biort='near_sym_a', qshift='qshift_a')
```

```python
x = torch.zeros(1,6,32,32)
yl, yh = xfm(x)
```

```python
yh[0].data[0,0,0,8,8,0] = 1
yh[0].data[0,1,1,8,8,0] = 1
yh[0].data[0,2,2,8,8,0] = 1
yh[0].data[0,3,3,8,8,0] = 1
yh[0].data[0,4,4,8,8,0] = 1
yh[0].data[0,5,5,8,8,0] = 1
```

```python
Y = ifm((yl, yh))
y = Y.data.numpy()
```

```python
fig, ax = plt.subplots(2,3)
for i in range(2):
    for j in range(3):
        ax[i][j].imshow(y[0,i*3+j])
```
